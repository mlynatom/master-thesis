### Detect languages to filter the czech ones

```python
from lingua import LanguageDetectorBuilder
detector = LanguageDetectorBuilder.from_all_languages().build()

detector.detect_multiple_languages_of('Hello, how are you?')
```

```python
texts = [" ".join(text["value"]) for text in dataset["train"]["conversations"]]
```

```python
texts[:3]
```

```python
languages = detector.detect_languages_in_parallel_of(texts)
```

```python
languages[:2]
```

```python
import numpy as np
iso_codes = np.array([lang.iso_code_639_1.name.lower() if lang is not None else "" for lang in languages])
```

```python
langs, counts = np.unique(iso_codes, return_counts=True)

for lang, count in zip(langs, counts):
    print(f"{lang}: {count}")
```

```python
indices = np.asarray(iso_codes == "cs").nonzero()
indices
```

```python
for index in indices[0]:
    print(texts[index])
    print("*"*100)
```

```python
indices
```

```python
#filter dataset to only include czech
czech_dataset = dataset["train"].select(list(indices[0]))
czech_dataset
```

```python
#filter rows with less than 2 turns in the conversation (human and gpt)
def filter_conversations(row):
    turns = row["conversations"]["from"]
    return len(turns) >= 2 and 'human' in turns and 'gpt' in turns

czech_dataset = czech_dataset.filter(filter_conversations)
czech_dataset
```

```python
from ftfy import fix_text
import re

def share_gpt_map(instances):
    old_conversations = instances["conversations"]
    turns = old_conversations["from"]
    texts = old_conversations["value"]

    if turns[0] != "human":
        raise ValueError("First turn should be human")
    
    if turns[1] != "gpt":
        raise ValueError("Second turn should be gpt")

    conversations = []
    for i in range(0, 2):
        text = fix_text(texts[i])
        text = re.sub(r'<.*?>+', " ", text).strip()
        text = text.replace("\n", " ")

        conversations.append({
            "role": "user" if i % 2 == 0 else "assistant",
            "content": text
        })

    instances["conversations"] = conversations
    return instances


czech_dataset = czech_dataset.map(share_gpt_map)
    
```
